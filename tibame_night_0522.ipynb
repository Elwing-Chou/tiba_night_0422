```python
# pandas的轉換操作
# apply(不加小括號的函式)
r = df["rating"] * 2
r.apply(round)
```

```python
def change(n):
    return round(n * 2, 1)
df["rating(10)"] = df["rating"].apply(change)
df
```

```python
key = ["星期一",
        "星期二",
        "星期三",
        "星期四",
        "星期五",
        "星期六",
        "星期日"]
def change(s):
    dic = {
        "星期一":1,
        "星期二":1,
        "星期三":1,
        "星期四":1,
        "星期五":1,
        "星期六":1,
        "星期日":1,
    }
    # 走過這七個字串, 檢查這些東西有沒有在原本裡面,
    # 如果有就是那天沒開
    for k in key:
        if k in s:
            dic[k] = 0
    return pd.Series(dic)
# pandas DataFrame: 表格型態(有行有列) Series: 一行或一列
df[key] = df["holiday"].apply(change)
df
```

```python
# 禮拜一或者禮拜二有開  並且 評分在3.8分以上
fil = (df["星期一"] == 1) | (df["星期二"] == 1)
fil = fil & (df["rating"] >= 3.8)
df[fil]
```
